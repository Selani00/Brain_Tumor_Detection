make all the data in the same name and same type

```python
folder = 'BarinTumorDataSet/yes/'

count =1 

for filename in os.listdir(folder):
    source = folder + filename
    destination = folder + "Y_" +str(count) + ".jpg"
    os.rename(source,destination)
    count = count + 1
print("Yes folder renamed")
```

```python
folder = 'BarinTumorDataSet/no/'

count =1 

for filename in os.listdir(folder):
    source = folder + filename
    destination = folder + "N_" +str(count) + ".jpg"
    os.rename(source,destination)
    count = count + 1
print("No folder renamed")
```

```python
# EDA(Exploratory Data Analysis)

listyes = os.listdir('BarinTumorDataSet/yes')
number_files_yes = len(listyes)
print("Number of Yes images: ",number_files_yes)


listno = os.listdir('BarinTumorDataSet/no')
number_files_no = len(listno)
print("Number of Yes images: ",number_files_no)
```

```python
# plot

data = {'tumorous' : number_files_yes, 'non-tumorous' : number_files_no}

typex = data.keys()
values = data.values()

fig = plt.figure(figsize =(10, 5))
plt.bar(typex, values, color ='maroon')

plt.xlabel("Data")
plt.ylabel("Number of Brain MRI images")
plt.title("Number of Brain MRI images in each category")

plt.show()
```
